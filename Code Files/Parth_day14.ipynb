```python
# Function to find defaulter percentage

def feature_defaulter_percentage(dataset,col):
    new_df = pd.DataFrame(columns=['Value','Defaulter percentage'])
    
    for feature in dataset[col].unique():
        default_per = dataset[(dataset[col]==feature) & (dataset.TARGET==1)].shape[0]
        total = dataset[dataset[col] == feature].shape[0]
        new_df = new_df.append({'Value':feature,'Defaulter percentage':(default_per*100/total)},ignore_index=True)
    return new_df.sort_values(by='Defaulter percentage',ascending=False)
```

```python
feature_defaulter_percentage(dataset = data,col = 'NAME_FAMILY_STATUS')
```
