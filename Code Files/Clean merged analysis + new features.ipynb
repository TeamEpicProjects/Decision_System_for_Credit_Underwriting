### New features

```python
# income:credt
```

```python
# 1 if income is less than credit and 0 if credit is more than income
conditions = [
    (data['AMT_INCOME_TOTAL'] < data['AMT_CREDIT_x']),
    (data['AMT_CREDIT_x'] > data['AMT_INCOME_TOTAL'])]

values = [0, 1]
```

```python
data['income VS. credit'] = np.select(conditions, values)
```

```python
# correct it
data['Family_type']=pd.cut(data['CNT_FAM_MEMBERS'],bins=[0, 4, 8, 16],labels=['Single','Nuclear family','Joint family'])
```

```python
data['Family_type'].value_counts()
```

```python
# Function to find defaulter percentage

def feature_defaulter_percentage(dataset,col):
    new_df = pd.DataFrame(columns=['Value','Defaulter percentage'])
    
    for feature in dataset[col].unique():
        default_per = dataset[(dataset[col]==feature) & (dataset.TARGET==1)].shape[0]
        total = dataset[dataset[col] == feature].shape[0]
        new_df = new_df.append({'Value':feature,'Defaulter percentage':(default_per*100/total)},ignore_index=True)
    return new_df.sort_values(by='Defaulter percentage',ascending=False)
```

```python
plt.figure(figsize = (13,6))
plt.subplot(1, 2, 1)
sns.countplot(data['Family_type'],hue=data['TARGET'])
plt.title("Applicants VS. Family types")


plt.subplot(1, 2, 2)
sns.countplot(Target1['Family_type'])
plt.title("Defaulters VS Family types")
plt.ticklabel_format(style='plain',axis='y')
plt.show()
```

- 98.66% of the applicants are single and around 98.43% of the single applicants are defaulters.
