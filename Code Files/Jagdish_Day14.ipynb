```python
def get_insight(col,dataframe):
  plt.figure(figsize=(20,10))
  dataframe[col]=dataframe[col].fillna(4)
  lst=list(dataframe[col].unique())
  if 4 in lst:
    lst.remove(4)
  dataframe[col]=dataframe[col].replace(to_replace =lst, value =1)
  lst.append(4)
  g =sns.scatterplot(x="SK_ID_CURR", y=col,
              hue=col,palette=['black','orange'],data=dataframe);
  g.set(xscale="log");
  return 0

get_insight("HOUSETYPE_MODE",df5)
  
```

```python
dict_nan=round(df.isna().sum().sort_values()/len(df["TARGET"])*100,2).to_dict()
print(dict_nan)
```

```python

#function for distribution of data ,wrt defaulter data,wrt whole data

def create_insight(col,df):
  print(f"NAN value percentage:{dict_nan[col]}")
  plt.figure(figsize=(10,10))
  index=df[col].value_counts().index
  value_d=round(df[col].value_counts()/len(df["TARGET"])*100)
  value_w=round(df[df["TARGET"]==1][col].value_counts()/len(df["TARGET"])*100)
  value_f=round(df[df["TARGET"]==1][col].value_counts()/len(df[df["TARGET"]==1])*100)


  X = index
  X_axis = np.arange(len(X))
    
  plt.bar(X_axis - 0.1, value_d, 0.1, label = 'distribution',color="blue")
  plt.bar(X_axis - 0, value_w, 0.1, label = 'defaulter w.r.t whole data',color="orange")
  plt.bar(X_axis + 0.1, value_f, 0.1, label = 'defaulter w.r.t defaulter data',color="red")
  
  plt.grid()
  plt.xticks(X_axis, X)
  plt.xlabel(col)
  plt.ylabel("percentage")
  plt.yticks(range(0,100,5))
  plt.legend()
  plt.show()

```
